## Import

```python
import pandas as pd
import numpy as np
```

## Read the dataset

```python
data = pd.read_csv("water_potability.csv")
```

```python
data
```

# Check missing value

```python
data.isnull().sum()
```

```python
data_cleaned = data.dropna()
```

```python
data_cleaned.isnull().sum()
```

## EDA

```python
data_cleaned.info()
```

```python
data_cleaned.describe()
```

## Calculate the count of potable (safe for consumption) and non-potable (unsafe for consumption)

```python
potable_count = data_cleaned['Potability'].value_counts()
potable_count
```

## Calculate the average pH value

```python
average_ph = data_cleaned['ph'].mean()
average_ph
```

## Calculate the correlation between features in the cleaned dataset

```python
correlation = data_cleaned.corr()
correlation
```

## Display the number of dropped rows

```python
rows_dropped = len(data) - len(data_cleaned)
print("Number of dropped rows:", rows_dropped)
```
